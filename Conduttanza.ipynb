```python
E = np.linspace(-5,5,500)
```

Calcola $G(E)$/$G_0$

```python
def g_e(E,E_M,gamma):
    return gamma**2*E**2/((E**2-E_M**2)**2+gamma**2*E**2)
```

```python
fig = plt.figure()
ax = plt.gca()
plt.plot(E, g_e(E, 0, 0.5), color = 'r')
plt.plot(E, g_e(E, 1, 0.5), color = 'b')
#plt.plot(E, 1/(E**2+1))
# plt.axhline(y=1, color='black', linestyle='--', linewidth=0.5) 
# plt.axvline(x=1, color='black', linestyle='--', linewidth=0.5) 
# plt.axvline(x=-1, color='black', linestyle='--', linewidth=0.5) 
# plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5) 
ax.set_xticks(ticks=[-1, 0, 1])
ax.set_ylim(-0.05,1.05)
ax.set_xlim(-4,4)
ax.set_xlabel('eV (unità arbitrarie)',size=18)
ax.set_ylabel(r'$G/G_0$',size=20)
plt.savefig('conduttanza', dpi=600)
plt.show()
```

Per $E_M=0$, in presenza di modi Majorana a zero energia, abbiamo un solo picco di conduttanza a zero bias con altezza quantizzata $G_0=2e^2/h$.
Per $E_M > 0$, in presenza di splitting, il picco a zero bias si divide in due picchi separati a $eV = \pm E_M=\pm 1$.
