```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

Calcola $I/I_C = \sin({\phi/2})$

```python
def corr(phi,nm):
    return np.sin(phi/2)*(2*nm-1)
```

```python
phi = np.linspace(0,4*np.pi,200)
```

```python
fig, ax = plt.subplots()
ax.plot(phi,corr(phi,0), color='r')
ax.plot(phi,corr(phi,1), color='b')
ax.set_xticks(ticks=[0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi], labels = ['0', '$\pi$', '$2\pi$','$3\pi$', '$4\pi$'])
ax.text(0.6, 0.8, r'$|1\rangle$', size=15)
ax.text(0.6, -0.9, r'$|0\rangle$', size=15)
ax.set_xlabel('$\phi$',size=20)
ax.set_ylabel('$I/I_C$',size=20)
plt.savefig('corrente di josephson', dpi =600)
plt.show()
```

Si nota un periodo di $4\pi$ della corrente di Josephson in presenza di modi Majorana con energia nulla.
