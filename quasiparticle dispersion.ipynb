```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
def ol(k, mu, b, delta, alpha, p):
    return (np.sqrt((k**2 - mu)**2 + b**2 + alpha**2*k**2 + delta**2 +p*2*np.sqrt((k**2 - mu)**2*(b**2 + alpha**2*k**2) + delta**2*b**2)))
```

```python
k = np.linspace(-np.pi, np.pi, 100)
b = 10**(-4)
mu = 0.1
alpha = 0 
delta = 0.2*10**(-3)
c = 1/(2*12000)
```

```python
Eap = np.sqrt((k**2 - mu)**2 + b**2 + delta**2 + 2*np.sqrt((k**2-mu)**2*b**2+delta**2*b**2))
Eam = np.sqrt((k**2 - mu)**2 + b**2 + delta**2 - 2*np.sqrt((k**2-mu)**2*b**2+delta**2*b**2))
```

```python
plt.plot(k,Eap)
plt.plot(k, -Eam)
plt.plot(k,-Eap)
plt.plot(k, +Eam)
#plt.plot(k,k**2 - mu + b, linestyle='--')
#plt.plot(k, k**2 - mu - b, linestyle='--')
```

```python
fig = plt.figure()
ax = plt.gca()
plt.plot(k, ol(k, 1, np.sqrt(2), 1, 1, 1))
plt.plot(k, ol(k, 1, np.sqrt(2), 1, 1, -1))
plt.plot(k, -ol(k, 1, np.sqrt(2), 1, 1, 1))
plt.plot(k, -ol(k, 1, np.sqrt(2), 1, 1, -1))
```
