La relazione di dispersione $E_k$ in rapporto all'hopping parameter $t$

$E_k/t = \sqrt{\mu^2/t^2 + 4cos(ka)^2 + 4 \mu/t cos(ka) + 4\Delta^2/t^2 sin(ka)^2}$

La funzione kitaev riceve in ingresso $\mu/t$, $\Delta/t$ e $k*a$

```python
def kitaev(mu, Delta = 1/2, k = np.linspace(-np.pi, np.pi, 100)):
    return np.sqrt((mu + 2*np.cos(k))**2 + 4*Delta**2*np.sin(k)**2)
```

```python
k = np.linspace(-np.pi, np.pi, 100)
Delta = 1/2
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8), sharex=True, sharey=True)

ax = axes[0][0]
ax.set_xticks(ticks=[-np.pi, 0, np.pi], labels = ['$-\pi$', '0', '$\pi$'])
#kitaev dispersion
ax.plot(k,kitaev(-2.5), color = 'r')
ax.plot(k, -kitaev(-2.5), color = 'r')
#bare electron dispersion
ax.plot(k,kitaev(Delta=0, mu=-2.5), '--', color = 'b', lw=1)
ax.plot(k, -kitaev(Delta=0, mu=-2.5), '--', lw=1, color = 'b')
ax.text(-0.6, 3.80, r'$\mu = -2.5t$', #transform=ax.transAxes,
        size=15)
ax.set_ylabel('$E/t$', size = 20)

ax = axes[0][1]
# ax.tick_params()
#kitaev dispersion
ax.plot(k,kitaev(-2), color = 'r')
ax.plot(k, -kitaev(-2), color = 'r')
#bare electron dispersion
ax.plot(k,kitaev(Delta=0, mu=-2), '--', color = 'b', lw=1)
ax.plot(k, -kitaev(Delta=0, mu=-2), '--', lw=1, color = 'b')
ax.text(-0.6, 3.80, '$\u03BC = -2t$', #transform=ax.transAxes,
        size=15)
ax.grid(True)

ax = axes[0][2]
#kitaev dispersion
ax.plot(k,kitaev(-1), color = 'r')
ax.plot(k, -kitaev(-1), color = 'r')
#bare electron dispersion
ax.plot(k,kitaev(Delta=0, mu=-1), '--', color = 'b', lw=1)
ax.plot(k, -kitaev(Delta=0, mu=-1), '--', lw=1, color = 'b')
ax.text(-0.6, 3.80, '$\u03BC = -t$', #transform=ax.transAxes,
        size=15)


ax = axes[1][0]
#kitaev dispersion
ax.plot(k,kitaev(1), color = 'r')
ax.plot(k, -kitaev(1), color = 'r')
#bare electron dispersion
ax.plot(k,kitaev(Delta=0, mu=1), '--', color = 'b', lw=1)
ax.plot(k, -kitaev(Delta=0, mu=1), '--', lw=1, color = 'b')
ax.text(-0.6, 3.80, '$\u03BC = t$', #transform=ax.transAxes,
        size=15)
ax.set_xlabel('$k$', size = 18)
ax.set_ylabel('$E/t$', size = 20)

ax = axes[1][1]
#kitaev dispersion
ax.plot(k,kitaev(2), color = 'r')
ax.plot(k, -kitaev(2), color = 'r')
#bare electron dispersion
ax.plot(k,kitaev(Delta=0, mu=2), '--', color = 'b', lw=1)
ax.plot(k, -kitaev(Delta=0, mu=2), '--', lw=1, color = 'b')
ax.text(-0.6, 2.20, '$\u03BC = 2t$', #transform=ax.transAxes,
        size=15)
ax.set_xlabel('$k$', size = 18)

ax = axes[1][2]
#kitaev dispersion
ax.plot(k,kitaev(2.5), color = 'r')
ax.plot(k, -kitaev(2.5), color = 'r')
#bare electron dispersion
ax.plot(k,kitaev(Delta=0, mu=2.5), '--', color = 'b', lw=1)
ax.plot(k, -kitaev(Delta=0, mu=2.5), '--', lw=1, color = 'b')
ax.text(-0.6, 2.20, '$\u03BC = 2.5t$', #transform=ax.transAxes,
        size=15)
ax.set_xlabel('$k$', size = 18)
# plt.savefig('kitaev-bare electron1.svg', dpi = 200)
plt.show()
```

Relazione di dispersione nel caso del toy model

$E_k/t = \sqrt{\left(k^2 a^2 - \mu/t\right)^2 + \Delta^2/t^2 * k^2}$

Variabili riscalate introducendo l'hopping parameter t. Dubbio su secondo pezzo per mancanza
di $a^2$, la variabile $k$  stata introdotta come $k*a$,  giusto lo stesso?

```python
def toy_model(mu, Delta = 1/2, k = np.linspace(-np.pi, np.pi, 100)):
    return np.sqrt((k**2 - mu)**2 + Delta**2*k**2)
```

```python
plt.plot(k,toy_model(0))
plt.plot(k, -toy_model(0))
```
