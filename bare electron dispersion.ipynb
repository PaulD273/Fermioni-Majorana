```python
b_c = 2
alpha_c = 5
k1 = np.linspace(-8, 8, 100)
```

```python
Ec1 = (k1**2) - mu + np.sqrt(b_c**2+alpha_c**2*k1**2)
Ec2 = (k1**2) - mu - np.sqrt(b_c**2+alpha_c**2*k1**2)
```

```python
fig, ax = plt.subplots()
ax.plot(k1,Ec1)
ax.plot(k1, Ec2)
# ax.grid(True, which='both')
seaborn.despine(ax=ax, offset=0)
ax.set_ylim(-10,20)
plt.arrow(-6.03, 8.85, -1, 0, width = 0.15, color = 'black', alpha = 0.5)
plt.arrow(6.03, 8.85, 1, 0, width = 0.15, color = 'black', alpha = 0.5)
ax.annotate('',
            xy=(0, 1.20), xycoords='data',
            xytext=(0, -3.17), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3))
ax.annotate('2b',
            xy=(0.43, -1.62), xycoords='data')
ax.set_xlabel('k')
ax.set_ylabel('E')
```

```python
np.min(Ec1-Ec2)
```

```python
b_d = 1.25
alpha_d = 1
```
