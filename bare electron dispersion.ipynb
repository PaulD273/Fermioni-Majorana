```python
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
k = np.linspace(-np.pi, np.pi, 100)
b = 1
mu = 1
alpha = 1 

mu_b=1
alpha_b =2
kb = np.linspace(-5, 5, 100)

b_c = 2 
alpha_c = 5 
mu_c = -2
k1 = np.linspace(-8, 8, 100)

b_d = 5
mu_d = -0.1
alpha_d = 2
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

ax = axes[0][0]
Ea1 = k**2 - mu + b
Ea2 = k**2 - mu - b
ax.plot(k,Ea1, color='r', zorder=2)
ax.plot(k, Ea2, zorder=2, color='blue')
ax.set_ylim(-2.2,7.8)
ax.arrow(-2.54, 4.80, 0, -1, width = 0.1, facecolor = 'silver', edgecolor='black', alpha = 1, zorder=3, transform = ax.transData)
ax.arrow(2.54, 4.8, 0, -1, width = 0.1, facecolor = 'silver', edgecolor='black', alpha = 1, zorder=3, transform = ax.transData)
# transform = ax.transAxes
ax.annotate('',
            xy=(0, 0.08), xycoords='data',
            xytext=(0, -2), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3), transform = ax.transData)
ax.annotate('2b',
            xy=(0.15, -1.2), xycoords='data', size=15, transform = ax.transData)
ax.set_ylabel('$E$', size = 20)
ax.set_yticks([-2, 0])

ax = axes[0][1]
Eb1 = kb**2 - mu_b + alpha_b*kb
Eb2 = kb**2 - mu_b - alpha_b*kb
ax.plot(kb,Eb1, color='r', zorder=2)
ax.plot(kb, Eb2, color='blue')
ax.set_xlim(-5,5)
ax.set_ylim(-3.5,7.8)
ax.arrow(-2.98, 3.8, -0.8, 0, width = 0.2, facecolor = 'silver', edgecolor='black', zorder =3,head_length=0.4, transform = ax.transData)
ax.arrow(2.98, 3.8, 0.8, 0, width = 0.2, facecolor = 'silver', edgecolor='black', zorder =3,head_length=0.4, transform = ax.transData)
ax.annotate('',
            xy=(2.6, 0.1), xycoords='data',
            xytext=(2.6, -2.1), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3), transform = ax.transData)
ax.annotate('$\mu + E_{SO}$',
            xy=(2.8, -1.2), xycoords='data', size=15, transform = ax.transData)
ax.annotate('',
            xy=(-1.1, -2.25), xycoords='data',
            xytext=(1.1, -2.25), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3), transform = ax.transData)
ax.annotate('$2k_{SO}$',
            xy=(-0.6, -3.1), xycoords='data', size=15, transform = ax.transData)
ax.set_yticks([-2, 0])
ax.set_xticks([-1, 0, 1])

ax = axes[1][0]
Ec1 = (k1**2) - mu_c + np.sqrt(b_c**2+alpha_c**2*k1**2)
Ec2 = (k1**2) - mu_c - np.sqrt(b_c**2+alpha_c**2*k1**2)
ax.plot(k1,Ec1, color='r', zorder=2)
ax.plot(k1, Ec2, color='blue')

ax.set_ylim(-5,15)
ax.arrow(-5.5, 9.1, -1.2, -0.3, width = 0.3, facecolor = 'silver', edgecolor='black', zorder=3, head_length=0.6)
ax.arrow(5.5, 9.1, 1.2, -0.3, width = 0.3, facecolor = 'silver', edgecolor='black', zorder=3, head_length=0.6)
ax.annotate('',
            xy=(0, -0.3), xycoords='data',
            xytext=(0, 4.2), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3))
ax.annotate('2b',
            xy=(0.43, 1.8), xycoords='data', size = 15)
ax.set_ylabel('$E$', size = 20)
ax.set_xlabel('$k$', size = 20)
ax.set_xticks([-2.5, 0, 2.5])
ax.set_yticks([-4, 0, 4])

ax = axes[1][1]
Ed1 = k1**2 - mu_d + np.sqrt(b_d**2+alpha_d**2*k1**2)
Ed2 = k1**2 - mu_d - np.sqrt(b_d**2+alpha_d**2*k1**2)
ax.set_ylim(-7,15)
ax.set_xlim(-2.5,2.5)
ax.plot(k,Ed1, color='r', zorder=2)
ax.plot(k, Ed2, color='blue')
ax.arrow(-1.6, 8.7, -0.08, -2.5, width = 0.07, facecolor = 'silver', edgecolor='black', zorder=3, head_length=1, transform = ax.transData)
ax.arrow(1.58, 8.7, 0.08, -2.5, width = 0.07, facecolor = 'silver', edgecolor='black', zorder=3, head_length=1, transform = ax.transData)
ax.annotate('',
            xy=(0, -5.2), xycoords='data',
            xytext=(0, 5.2), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3.5), transform = ax.transData)
ax.annotate('2b',
            xy=(0.17,0.5), xycoords='data', size = 17, transform = ax.transData)
ax.set_xlabel('$k$', size = 20)
ax.set_yticks([-5, 0, 5])
# plt.savefig('bare_el_disp_nano3.png', dpi=600)
plt.show()
```

# Figura 2

$\mu=1$, $\alpha=2$ e $b=0$

Il punto di minimo è a $\mu + E_{SO}$ al di sotto dello zero. Dato che $E_{SO}=m\alpha^2/2 \hbar^2$, e che è stato posto per semplicità $\hbar^2/2m=1$ e $\alpha=2$, $E_{SO}=\alpha^2/4=1$ e $\mu=1$.

```python
0-np.min(Eb1)
```

La distanza in $x$ tra i due minimi è $2 k_{SO}= 2m\alpha/\hbar^2=\alpha=2$

```python
i=np.argmin(Eb1)
i2=np.argmin(Eb2)
i,i2
```

```python
-kb[i]+kb[i2]
```

# Figura 3

$\mu=-2$, $b=2$ e $\alpha=5$

Distanza $2b = 4$ tra le due figure per $k=0$ 

```python
ic1 = np.argmin(Ec1)
ic2 = np.argmin(Ec2)
np.argmin(Ec1), np.argmin(Ec2)
```

```python
np.min(Ec1)+np.max(Ec2[ic1:ic2])
```

I punti di minimo si trovano a $k = \pm \sqrt{k^2_{SO} - k_Z^4/(4k^4_{SO})} = \pm \sqrt{(\alpha/2)^2-b^2/\alpha^2} = \pm \sqrt{(5/2)^2-4/25}\approx \pm 2.467$

```python
k1[ic2]
```
