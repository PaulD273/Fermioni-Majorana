```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
k = np.linspace(-np.pi, np.pi, 100)
b = 1
mu = 1
alpha = 1 
```

```python
Ea1 = k**2 - mu + b
Ea2 = k**2 - mu - b
```

```python
plt.plot(k,Ea1)
plt.plot(k, Ea2)
```

```python
Eb1 = k**2 - mu + alpha*k
Eb2 = k**2 - mu - alpha*k
```

```python
plt.plot(k,Eb1)
plt.plot(k, Eb2)
plt.xlim(-4,4)
```

```python
b_c = 1
alpha_c = 4
```

```python
Ec1 = k**2 - mu + np.sqrt(b_c**2+alpha_c**2*k**2)
Ec2 = k**2 - mu - np.sqrt(b_c**2+alpha_c**2*k**2)
```

```python
fig, ax = plt.subplots()
ax.plot(k,Ec1)
ax.plot(k, Ec2)
ax.grid(True, which='both')
seaborn.despine(ax=ax, offset=0)
```

```python
b_d = 1.25
alpha_d = 1
```

```python
Ed1 = k**2 - mu + np.sqrt(b_d**2+alpha_d**2*k**2)
Ed2 = k**2 - mu - np.sqrt(b_d**2+alpha_d**2*k**2)
```

```python
fig, ax = plt.subplots()
ax.plot(k,Ed1)
ax.plot(k, Ed2)
ax.grid(True, which='both')
seaborn.despine(ax=ax, offset=0)
```
