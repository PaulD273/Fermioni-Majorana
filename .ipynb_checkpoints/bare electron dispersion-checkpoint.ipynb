```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
%matplotlib widget
```

```python
k = np.linspace(-np.pi, np.pi, 100)
b = 1
mu = 1
alpha = 1 
```

```python
Ea1 = k**2 - mu + b
Ea2 = k**2 - mu - b
```

```python
plt.plot(k,Ea1)
plt.plot(k, Ea2)
```

```python
Eb1 = k**2 - mu + alpha*k
Eb2 = k**2 - mu - alpha*k
```

```python
ax = plt.gca()
plt.plot(k,Eb1)
plt.plot(k, Eb2)
plt.xlim(-4,4)
plt.ylim(-1.5,5)
```

```python
np.min(Eb1) 
```

```python
b_c = 2
alpha_c = 5
k1 = np.linspace(-8, 8, 100)
```

```python
Ec1 = (k1**2) - mu + np.sqrt(b_c**2+alpha_c**2*k1**2)
Ec2 = (k1**2) - mu - np.sqrt(b_c**2+alpha_c**2*k1**2)
```

```python
fig, ax = plt.subplots()
ax.plot(k1,Ec1)
ax.plot(k1, Ec2)
# ax.grid(True, which='both')
seaborn.despine(ax=ax, offset=0)
ax.set_ylim(-10,20)
plt.arrow(-6.03, 8.85, -1, 0, width = 0.15, color = 'black', alpha = 0.5)
plt.arrow(6.03, 8.85, 1, 0, width = 0.15, color = 'black', alpha = 0.5)
ax.annotate('',
            xy=(0, 1.20), xycoords='data',
            xytext=(0, -3.17), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10"))
ax.annotate('2b',
            xy=(0.43, -1.62), xycoords='data')
ax.set_xlabel('k')
ax.set_ylabel('E')
```

```python
np.min(Ec1-Ec2)
```

```python
b_d = 1.25
alpha_d = 1
```

```python
Ed1 = k**2 - mu + np.sqrt(b_d**2+alpha_d**2*k**2)
Ed2 = k**2 - mu - np.sqrt(b_d**2+alpha_d**2*k**2)
```

```python
fig, ax = plt.subplots()
ax.plot(k,Ed1)
ax.plot(k, Ed2)
ax.grid(True, which='both')
seaborn.despine(ax=ax, offset=0)
```

```python
np.min(Ed1-Ed2)
```
