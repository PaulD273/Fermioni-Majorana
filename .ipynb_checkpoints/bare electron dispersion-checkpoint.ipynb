```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
k = np.linspace(-np.pi, np.pi, 100)
b = 1
mu = 1
alpha = 1 
b_c = 2
alpha_c = 5
mu_c = -2
k1 = np.linspace(-8, 8, 100)
b_d = 1.25
mu_d = -2
alpha_d = 1
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8))

ax = axes[0][0]
Ea1 = k**2 - mu + b
Ea2 = k**2 - mu - b
ax.plot(k,Ea1, color='r', zorder=2)
ax.plot(k, Ea2, zorder=2)
ax.set_ylim(-2.2,7.8)
ax.arrow(-2.54, 4.80, 0, -0.8, width = 0.06, color = 'black', alpha = 1, zorder=3)
ax.arrow(2.54, 4.8, 0, -0.8, width = 0.06, color = 'black', alpha = 1, zorder=3)
# transform = ax.transAxes
ax.annotate('',
            xy=(0, 0.08), xycoords='data',
            xytext=(0, -2.09), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3))
ax.annotate('2b',
            xy=(0.15, -1.2), xycoords='data', size=15)

ax = axes[0][1]
Eb1 = k**2 - mu + alpha*k
Eb2 = k**2 - mu - alpha*k
ax.plot(k,Eb1, color='r', zorder=2)
ax.plot(k, Eb2)
ax.set_xlim(-4,4)
ax.set_ylim(-2.4,5)
ax.arrow(-2.11, 2.47, -0.45, 0, width = 0.08, color = 'black', zorder =3)
ax.arrow(2.11, 2.47, 0.45, 0, width = 0.08, color = 'black', zorder =3)
ax.annotate('',
            xy=(1.73, 0.1), xycoords='data',
            xytext=(1.73, -1.33), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3))
ax.annotate('$E_{SO}$',
            xy=(1.9, -0.78), xycoords='data', size=15)
ax.annotate('',
            xy=(-0.62, -1.38), xycoords='data',
            xytext=(0.62, -1.38), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3))
ax.annotate('$2k_{SO}$',
            xy=(-0.3, -2), xycoords='data', size=15)


ax = axes[1][0]
Ec1 = (k1**2) - mu_c + np.sqrt(b_c**2+alpha_c**2*k1**2)
Ec2 = (k1**2) - mu_c - np.sqrt(b_c**2+alpha_c**2*k1**2)
ax.plot(k1,Ec1, color='r', zorder=2)
ax.plot(k1, Ec2)

ax.set_ylim(-5,15)
ax.arrow(-5.5, 9.1, -1, -0.3, width = 0.3, color = 'black', zorder=3, head_length=0.6)
ax.arrow(5.5, 9.1, 1, -0.3, width = 0.3, color = 'black', zorder=3, head_length=0.6)
ax.annotate('',
            xy=(0, -0.4), xycoords='data',
            xytext=(0, 4.4), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                    connectionstyle="angle,angleA=0,angleB=-90,rad=10", linewidth=3))
ax.annotate('2b',
            xy=(0.43, 1.8), xycoords='data', size = 15)


ax = axes[1][1]
Ed1 = k**2 - mu_d + np.sqrt(b_d**2+alpha_d**2*k**2)
Ed2 = k**2 - mu_d - np.sqrt(b_d**2+alpha_d**2*k**2)
ax.plot(k,Ed1, color='r', zorder=2)
ax.plot(k, Ed2)
plt.show()
```
