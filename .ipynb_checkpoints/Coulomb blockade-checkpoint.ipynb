```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_g = np.linspace(-2,2,100)
```

Calcolo di $E(N)/E_C$

```python
def en(N,delta):
    return (N-n_g)**2 +delta
```

```python
fig = plt.figure()
ax = plt.gca()
plt.plot(n_g,en(0,0))
plt.plot(n_g,en(1,0))
plt.plot(n_g,en(-1,0))
plt.plot(n_g,en(2,0))
plt.plot(n_g,en(-2,0))
ax.set_ylim(0,1.5)
ax.set_xlim(-2,2)
```

```python
fig = plt.figure()
ax = plt.gca()
plt.plot(n_g,en(0,0))
plt.plot(n_g,en(1,1.2))
plt.plot(n_g,en(-1,1.2))
plt.plot(n_g,en(2,0))
plt.plot(n_g,en(-2,0))
ax.set_ylim(0,1.5)
ax.set_xlim(-2,2)
```

```python
fig = plt.figure()
ax = plt.gca()
plt.plot(n_g,en(0,0))
plt.plot(n_g,en(1,0.7))
plt.plot(n_g,en(-1,0.7))
plt.plot(n_g,en(2,0))
plt.plot(n_g,en(-2,0))
ax.set_ylim(0,1.5)
ax.set_xlim(-2,2)
```

```python
def g_e(E,E_M,gamma):
    return gamma**2*E**2/((E**2-E_M**2)**2+gamma**2*E**2)
```
