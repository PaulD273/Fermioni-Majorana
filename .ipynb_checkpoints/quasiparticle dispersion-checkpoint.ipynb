```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
k = np.linspace(-np.pi, np.pi, 100)
b = 1
mu = 1
alpha = 1 
delta = 1
```

```python
Eap = np.sqrt((k**2 - mu)**2 + b**2 + delta**2 + 2*np.sqrt((k**2-mu)**2*b**2+delta**2*b**2))
Eam = np.sqrt((k**2 - mu)**2 + b**2 + delta**2 - 2*np.sqrt((k**2-mu)**2*b**2+delta**2*b**2))
```

```python
plt.plot(k,Eap)
plt.plot(k, -Eam)
plt.plot(k,-Eap)
plt.plot(k, +Eam)
```

```python
Eap = np.sqrt((k**2 - mu)**2 + b**2 + alpha**2*k**2+ delta**2 + 2*np.sqrt((k**2-mu)**2*(b**2 + alpha**2*k**2)+delta**2*b**2))
Eam = np.sqrt((k**2 - mu)**2 + b**2 + alpha**2*k**2+ delta**2 - 2*np.sqrt((k**2-mu)**2*(b**2 + alpha**2*k**2)+delta**2*b**2))
```

```python
plt.plot(k,Eap)
plt.plot(k, -Eam)
plt.plot(k,-Eap)
plt.plot(k, +Eam)
```
