```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%matplotlib widget
```

```python
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
angles = np.linspace(0 * np.pi, 2 * np.pi, 100 )
r = 1
xs = r * np.cos(angles)
ys = r * np.sin(angles)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16,5))

ax = axes[0]
ax.plot(xs, ys, color = 'blue')
ax.spines[['left', 'bottom']].set_position('center')
ax.spines[['top', 'right']].set_visible(False)
# ax.set_xticks([])
# ax.set_yticks([])
ax.set_xlabel('$h_y$', labelpad=10, loc='right', size=15, rotation='horizontal')
ax.set_ylabel('$h_z$', labelpad=0, size=15, loc='top', rotation='horizontal')

ax = axes[1]
ax.plot(xs+1, ys+1, color = 'blue')
ax.spines.left.set_position(('data', 1.0))
ax.spines.bottom.set_position( ('data', 0.0))
ax.spines[['top', 'right']].set_visible(False)
# ax.set_xticks([])
# ax.set_yticks([])
ax.set_xlabel('$h_y$', labelpad=10, loc='right', size=15, rotation='horizontal')
ax.set_ylabel('$h_z$', labelpad=0, size=15, loc='top', rotation='horizontal')
#plt.savefig('cerchio', bbox_inches='tight')
plt.show()
```
