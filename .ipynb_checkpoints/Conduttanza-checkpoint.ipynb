```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
```

```python
E = np.linspace(-5,5,200)
```

Calcola $G(E)$/$G_0$

```python
def g_e(E,E_M,gamma):
    return gamma**2*E**2/((E**2-E_M**2)**2+gamma**2*E**2)
```

```python
fig = plt.figure()
ax = plt.gca()
plt.plot(E, g_e(E, 0, 1))
plt.plot(E, g_e(E, 1, 1))
#plt.plot(E, 1/(E**2+1))
plt.axhline(y=1, color='black', linestyle='--', linewidth=0.5) 
plt.axvline(x=1, color='black', linestyle='--', linewidth=0.5) 
plt.axvline(x=-1, color='black', linestyle='--', linewidth=0.5) 
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5) 
ax.set_ylim(0,1.05)
ax.set_xlim(-4,4)
ax.set_xlabel('eV (unit arbitrarie)')
ax.set_ylabel(r'G/$G_0$')
```
