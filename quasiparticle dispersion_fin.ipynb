```python
k = np.linspace(-3/2*np.pi, 3/2*np.pi, 200)#np.linspace(-np.pi, np.pi, 200)
mu_a = 0
b_a=1
delta_a = 0.5

k2 = np.linspace(-3/2*np.pi, 3/2*np.pi, 200)
mu_b = 0
alpha_b=2
delta_b = 1

mu_c =0
alpha_c=1.5
delta_c = 0.5
b_c = 0.2#0.8
kc = np.linspace(-3/2*np.pi, 3/2*np.pi, 200)

mu_d =0#3
alpha_d=3#1.5
delta_d = 1.5# 2

mu_e =0#3
alpha_e=3#1.5
delta_e = 1.5#2

kf = np.linspace(-2*np.pi, 2*np.pi, 200)
mu_f =0
alpha_f=3
delta_f = 1.5
```

```python
fig, axes = plt.subplots(2, 3, figsize=(2*16,2*9), sharey = True)

ax = axes[0][0]
Ea1 = np.sqrt((k**2-mu_a)**2 + b_a**2 + delta_a**2 + 2*np.sqrt((k**2-mu_a)**2*b_a**2 + delta_a**2*b_a**2))
Ea2 = -np.sqrt((k**2-mu_a)**2 + b_a**2 + delta_a**2 + 2*np.sqrt((k**2-mu_a)**2*b_a**2 + delta_a**2*b_a**2))
Ea3 = np.sqrt((k**2-mu_a)**2 + b_a**2 + delta_a**2 - 2*np.sqrt((k**2-mu_a)**2*b_a**2 + delta_a**2*b_a**2))
Ea4 = -np.sqrt((k**2-mu_a)**2 + b_a**2 + delta_a**2 - 2*np.sqrt((k**2-mu_a)**2*b_a**2 + delta_a**2*b_a**2))

ax.plot(k, Ea1, color = 'red', linewidth = 3)
ax.plot(k, Ea2, color = 'red', linewidth = 3)
ax.plot(k, Ea3, color = 'red', linewidth = 3)
ax.plot(k, Ea4, color = 'red', linewidth = 3)

ax.plot(k, k**2 - mu_a + b_a, '--', color = 'red')
ax.plot(k, k**2 - mu_a - b_a, '--', color = 'red')
ax.plot(k, -(k**2 - mu_a + b_a), '--', color = 'red')
ax.plot(k, -(k**2 - mu_a - b_a), '--', color = 'red')
#ax.set_ylim(-12,12)
ax.set_xticks([0])
ax.set_yticks([0])
ax.set_ylabel('E', size=30)

ax.set_title('(a)', loc='left', size = 30)

ax = axes[0][1]
Eb1 = np.sqrt((k2**2-mu_b)**2 + alpha_b**2*k2**2 + delta_b**2 + 2*np.sqrt((k2**2-mu_b)**2*alpha_b**2*k2**2))
Eb2 = -np.sqrt((k2**2-mu_b)**2 + alpha_b**2*k2**2 + delta_b**2 + 2*np.sqrt((k2**2-mu_b)**2*alpha_b**2*k2**2))
Eb3 = np.sqrt((k2**2-mu_b)**2 + alpha_b**2*k2**2 + delta_b**2 - 2*np.sqrt((k2**2-mu_b)**2*alpha_b**2*k2**2))
Eb4 = -np.sqrt((k2**2-mu_b)**2 + alpha_b**2*k2**2 + delta_b**2 - 2*np.sqrt((k2**2-mu_b)**2*alpha_b**2*k2**2))

ax.plot(k2, Eb1, color = '#ff7f0e', linewidth = 3)
ax.plot(k2, Eb2, color = '#ff7f0e', linewidth = 3)
ax.plot(k2, Eb3, color = '#ff7f0e', linewidth = 3)
ax.plot(k2, Eb4, color = '#ff7f0e', linewidth = 3)

ax.plot(k2, k2**2 - mu_b + alpha_b*k2, '--', color = '#ff7f0e')
ax.plot(k2, k2**2 - mu_b - alpha_b*k2, '--', color = '#ff7f0e')
ax.plot(k2, -(k2**2 - mu_b + alpha_b*k2), '--', color = '#ff7f0e')
ax.plot(k2, -(k2**2 - mu_b - alpha_b*k2), '--', color = '#ff7f0e')
ax.set_ylim(-12,12)
ax.set_xticks([0])
ax.set_yticks([0])
ax.set_title('(b)', loc='left', size = 30)

ax = axes[0][2]
Ec1 = np.sqrt((kc**2-mu_c)**2 + b_c**2 + alpha_c**2*kc**2 + delta_c**2 + 2*np.sqrt((kc**2-mu_c)**2*(alpha_c**2*kc**2 + b_c**2) + delta_c**2*b_c**2))
Ec2 = -np.sqrt((kc**2-mu_c)**2 + b_c**2 + alpha_c**2*kc**2 + delta_c**2 + 2*np.sqrt((kc**2-mu_c)**2*(alpha_c**2*kc**2 + b_c**2) + delta_c**2*b_c**2))
Ec3 = np.sqrt((kc**2-mu_c)**2 + b_c**2 + alpha_c**2*kc**2 + delta_c**2 - 2*np.sqrt((kc**2-mu_c)**2*(alpha_c**2*kc**2 + b_c**2) + delta_c**2*b_c**2))
Ec4 = -np.sqrt((kc**2-mu_c)**2 + b_c**2 + alpha_c**2*kc**2 + delta_c**2 - 2*np.sqrt((kc**2-mu_c)**2*(alpha_c**2*kc**2 + b_c**2) + delta_c**2*b_c**2))

ax.plot(kc, Ec1, color = '#e377c2', linewidth = 3)
ax.plot(kc, Ec2, color = '#e377c2', linewidth = 3)
ax.plot(kc, Ec3, color = '#e377c2', linewidth = 3)
ax.plot(kc, Ec4, color = '#e377c2', linewidth = 3)


ax.plot(kc, kc**2 - mu_c + np.sqrt(b_c**2+alpha_c**2*kc**2), '--', color = '#e377c2')
ax.plot(kc, kc**2 - mu_c - np.sqrt(b_c**2+alpha_c**2*kc**2), '--', color = '#e377c2')
ax.plot(kc, -(kc**2 - mu_c + np.sqrt(b_c**2+alpha_c**2*kc**2)), '--', color = '#e377c2')
ax.plot(kc, -(kc**2 - mu_c - np.sqrt(b_c**2+alpha_c**2*kc**2)), '--', color = '#e377c2')

ax.set_xticks([0])
ax.set_yticks([0])
ax.set_title('(c)', loc='left', size = 30)

ax = axes[1][0]
# Ef1 = np.sqrt((kc**2-mu_e)**2 + 2*(mu_e**2 + delta_e**2) + alpha_e**2*kc**2 + delta_e**2 + 2*np.sqrt((kc**2-mu_e)**2*(alpha_e**2*kc**2 + 2*(mu_e**2 + delta_e**2)) + delta_e**2*2*(mu_e*2 + delta_e**2)))
# Ef2 = -np.sqrt((kc**2-mu_e)**2 + 2*(mu_e**2 + delta_e**2) + alpha_e**2*kc**2 + delta_e**2 + 2*np.sqrt((kc**2-mu_e)**2*(alpha_e**2*kc**2 + 2*(mu_e**2 + delta_e**2)) + delta_e**2*2*(mu_e*2 + delta_e**2)))
Ef3 = np.sqrt((kf**2-mu_f)**2 + 0.5*(mu_f**2 + delta_f**2) + alpha_f**2*kf**2 + delta_f**2 - 2*np.sqrt((kf**2-mu_f)**2*(alpha_f**2*kf**2 + 0.5*(mu_f**2 + delta_f**2)) + delta_f**2*0.5*(mu_f*2 + delta_f**2)))
Ef4 = -np.sqrt((kf**2-mu_f)**2 + 0.5*(mu_f**2 + delta_f**2) + alpha_f**2*kf**2 + delta_f**2 - 2*np.sqrt((kf**2-mu_f)**2*(alpha_f**2*kf**2 + 0.5*(mu_f**2 + delta_f**2)) + delta_f**2*0.5*(mu_f*2 + delta_f**2)))

ax.plot(kf, Ef3, color = '#9467bd', linewidth = 3)
ax.plot(kf, Ef4, color = '#9467bd', linewidth = 3)

#ax.plot(kf, kf**2 - mu_f + np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2), '--', color = '#e377c2')
ax.plot(kf, kf**2 - mu_f - np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2), '--', color = '#9467bd')
#ax.plot(kf, -(kf**2 - mu_f + np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2)), '--', color = '#e377c2')
ax.plot(kf, -(kf**2 - mu_f - np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2)), '--', color = '#9467bd')

ax.set_xlim(-6,6)
ax.set_xticks([0])
ax.set_yticks([0])
ax.set_ylabel('E', size=30 )
ax.set_xlabel('k', size=30 )
ax.set_title('(d)', loc='left', size = 30)

ax = axes[1][1]
# Ed1 = np.sqrt((kc**2-mu_d)**2 + mu_d**2 + delta_d**2 + alpha_d**2*kc**2 + delta_d**2 + 2*np.sqrt((kc**2-mu_d)**2*(alpha_d**2*kc**2 + mu_d**2 + delta_d**2) + delta_d**2*(mu_d**2 + delta_d**2)))
# Ed2 = -np.sqrt((kc**2-mu_d)**2 + mu_d**2 + delta_d**2 + alpha_d**2*kc**2 + delta_d**2 + 2*np.sqrt((kc**2-mu_d)**2*(alpha_d**2*kc**2 + mu_d**2 + delta_d**2) + delta_d**2*(mu_d**2 + delta_d**2)))
Ed3 = np.sqrt((kf**2-mu_d)**2 + mu_d**2 + delta_d**2 + alpha_d**2*kf**2 + delta_d**2 - 2*np.sqrt((kf**2-mu_d)**2*(alpha_d**2*kf**2 + mu_d**2 + delta_d**2) + delta_d**2*(mu_d**2 + delta_d**2)))
Ed4 = -np.sqrt((kf**2-mu_d)**2 + mu_d**2 + delta_d**2 + alpha_d**2*kf**2 + delta_d**2 - 2*np.sqrt((kf**2-mu_d)**2*(alpha_d**2*kf**2 + mu_d**2 + delta_d**2) + delta_d**2*(mu_d**2 + delta_d**2)))

#ax.plot(kc, Ed1, color = '#17becf', alpha = 0.8)
#ax.plot(kc, Ed2, color = '#17becf', alpha = 0.8)
ax.plot(kf, Ed3, color = '#17becf', linewidth = 3)
ax.plot(kf, Ed4, color = '#17becf', linewidth = 3)

#ax.plot(kf, kf**2 - mu_f + np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2), '--', color = '#e377c2')
ax.plot(kf, kf**2 - mu_d - np.sqrt((mu_d**2 + delta_d**2)+alpha_d**2*kf**2), '--', color = '#17becf')
#ax.plot(kf, -(kf**2 - mu_f + np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2)), '--', color = '#e377c2')
ax.plot(kf, -(kf**2 - mu_d - np.sqrt((mu_d**2 + delta_d**2)+alpha_d**2*kf**2)), '--', color = '#17becf')

ax.set_xlim(-6,6)
ax.set_xticks([0])
ax.set_yticks([0])
ax.set_xlabel('k', size=30 )
ax.set_title('(e)', loc='left', size = 30)

ax = axes[1][2]
# Ee1 = np.sqrt((kc**2-mu_e)**2 + 2*(mu_e**2 + delta_e**2) + alpha_e**2*kc**2 + delta_e**2 + 2*np.sqrt((kc**2-mu_e)**2*(alpha_e**2*kc**2 + 2*(mu_e**2 + delta_e**2)) + delta_e**2*2*(mu_e*2 + delta_e**2)))
# Ee2 = -np.sqrt((kc**2-mu_e)**2 + 2*(mu_e**2 + delta_e**2) + alpha_e**2*kc**2 + delta_e**2 + 2*np.sqrt((kc**2-mu_e)**2*(alpha_e**2*kc**2 + 2*(mu_e**2 + delta_e**2)) + delta_e**2*2*(mu_e*2 + delta_e**2)))
Ee3 = np.sqrt((kf**2-mu_e)**2 + 2*(mu_e**2 + delta_e**2) + alpha_e**2*kf**2 + delta_e**2 - 2*np.sqrt((kf**2-mu_e)**2*(alpha_e**2*kf**2 + 2*(mu_e**2 + delta_e**2)) + delta_e**2*2*(mu_e*2 + delta_e**2)))
Ee4 = -np.sqrt((kf**2-mu_e)**2 + 2*(mu_e**2 + delta_e**2) + alpha_e**2*kf**2 + delta_e**2 - 2*np.sqrt((kf**2-mu_e)**2*(alpha_e**2*kf**2 + 2*(mu_e**2 + delta_e**2)) + delta_e**2*2*(mu_e*2 + delta_e**2)))

ax.plot(kf, Ee3, color = '#9467bd', linewidth = 3)
ax.plot(kf, Ee4, color = '#9467bd', linewidth = 3)

#ax.plot(kf, kf**2 - mu_f + np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2), '--', color = '#e377c2')
ax.plot(kf, kf**2 - mu_e - np.sqrt(2*(mu_e**2 + delta_e**2)+alpha_e**2*kf**2), '--', color = '#9467bd')
#ax.plot(kf, -(kf**2 - mu_f + np.sqrt(0.5*(mu_f**2 + delta_f**2)+alpha_f**2*kf**2)), '--', color = '#e377c2')
ax.plot(kf, -(kf**2 - mu_e - np.sqrt(2*(mu_e**2 + delta_e**2)+alpha_e**2*kf**2)), '--', color = '#9467bd')

ax.set_xlim(-6,6)
ax.set_xticks([0])
# plt.xticks(fontsize = 20)
ax.set_yticks([0])
ax.set_xlabel('k', size=30 )
ax.set_title('(f)', loc='left', size = 30)

plt.savefig('disp_ol_200.png', dpi = 200, bbox_inches = 'tight') 

plt.show()
```
